```python
# Import required modules
import pandas as pd
from sklearn import preprocessing

# Set charts to view inline
%matplotlib inline
```

```python
data = pd.read_csv('patch_level_radiomics_feature_TCGA-AO-A0J4-01Z-00-DX1.csv')

# cols
dat1 = data[['case_id', 'image_width', 'image_height', 'patch_x', 'patch_y', 'patch_width', 'patch_height', 'nuclei_ratio', 'nuclei_average_area', 'nuclei_average_perimeter']]
dat1
```

```python
cols = list(dat1.columns)
cols[8]
```

```python
dat2 = dat1[cols[8:]]
vals = dat2.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(vals)
dat2 = pd.DataFrame(x_scaled)
dat2.columns = ['nuclei_average_area', 'nuclei_average_perimeter']
dat2
```

```python
dat1 = dat1.drop(['nuclei_average_area', 'nuclei_average_perimeter'], axis=1)
dat1
# dat1.join(dat2)
# pd.concat([dat1, dat2], axis=1)
```

```python
df_normalized = pd.concat([dat1, dat2], axis=1)
```
